```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
```

```python
def plot_clusters(X, centroids, labels, title,prev_centroid):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
    plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='x')
    plt.scatter(prev_centroid[:, 0], prev_centroid[:, 1], s=300, c='gray', marker='x')
    plt.title(title)
    plt.show()
```

```python
pca = PCA(2)
X_pca = pca.fit_transform(X)
```

```python
np.random.seed(42)
```

```python
initial_centroids = X_pca[np.random.choice(X_pca.shape[0],3,replace=False)]
print(initial_centroids)
```

```python
kmeans = KMeans(n_clusters=3,init=initial_centroids,n_init=1,max_iter=1000,random_state=42,)
kmeans.fit(X_pca)
plot_clusters(X_pca, kmeans.cluster_centers_, kmeans.labels_, "First Iteration - Kmeans",initial_centroids)
```

```python
kmeans = KMeans(n_clusters=3,init=initial_centroids,n_init=1,max_iter=500,random_state=42,verbose=10)
kmeans.fit(X_pca)
plot_clusters(X_pca, kmeans.cluster_centers_, kmeans.labels_, "First Iteration - Kmeans",initial_centroids)
```

```python
np.random.seed(42)
initial_centroids = X_pca[np.random.choice(X_pca.shape[0], 3, replace=False)]
print("Initial Centroids:\n", initial_centroids)

kmeans = KMeans(n_clusters=3, init=initial_centroids, n_init=1, max_iter=1, random_state=42)
kmeans.fit(X_pca)
plot_clusters(X_pca, kmeans.cluster_centers_, kmeans.labels_, "First Iteration: K-Means",initial_centroids)
```

```python
old_cluster_centre = kmeans.cluster_centers_

kmeans = KMeans(n_clusters=3,init=old_cluster_centre,n_init=1,max_iter=1,random_state=42,)
kmeans.fit(X_pca)
plot_clusters(X_pca, kmeans.cluster_centers_, kmeans.labels_, "First Iteration - Kmeans",old_cluster_centre)
```

```python
wcss = []

for k in range(1,15):
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure()
plt.plot(range(1,15),wcss,marker='x',color='red')
plt.show()
```
